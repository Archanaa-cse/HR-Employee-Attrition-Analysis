```python
import re

def to_snake_case(col):
    col = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', col)
    col = re.sub('([a-z0-9])([A-Z])', r'\1_\2', col)
    return col.lower()

df.columns = [to_snake_case(c) for c in df.columns]
```

```python
df.columns
```

```python
df.info()
```

```python
drop_cols=['daily_rate','employee_count','hourly_rate','monthly_rate','over18','standard_hours']
df=df.drop(columns=drop_cols)
```

```python
df.columns
```

```python
df=df.rename(columns={'employee_number' : 'employee_id'})
```

```python
df.columns
```

```python
df['attrition'] = df['attrition'].map({'Yes': 1, 'No': 0})
```

```python
df['attrition'].value_counts(normalize=True) * 100
```

```python
df['over_time'] = df['over_time'].map({'Yes': 1, 'No': 0})
```

```python
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
# create category for working yrs
def tenurebucket(years):
    if years <= 2:
        return '0-2'
    elif years <= 5:
        return '3-5'
    elif years <= 10:
        return '6-10'
    else:
        return '10+'

df['tenure_bucket'] = df['years_at_company'].apply(tenurebucket)
```

```python
#create label for salary
df['income_level'] = pd.qcut(
    df['monthly_income'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
df['promotion_gap'] = df['years_at_company'] - df['years_since_last_promotion']
```

```python
df[['years_at_company','years_since_last_promotion','promotion_gap','income_level','tenure_bucket']].head()
```

```python
df['work_life_risk']=((df['over_time']==1) & (df['work_life_balance']<=2)).astype(int)
```

```python
df[['income_level','tenure_bucket','work_life_risk']].head()
```

```python
tenure_attrition = ((
    df.groupby('tenure_bucket')['attrition']
      .mean() * 100).round(2)
      .sort_values(ascending=False)
)

tenure_attrition
```

```python
income_attrition = ((
    df.groupby('income_level')['attrition']
      .mean() * 100).round(2)
      .sort_values(ascending=False)
)

income_attrition
```

```python
(df.groupby('over_time')['attrition'].mean() *100).round(2)
```

```python
(df.groupby(pd.cut(df['promotion_gap'], bins=[0,2,5,10,40]))['attrition'].mean() *100).round(2)
```
